```python
def get_result(url):
    rs = requests.session()
    #rs.proxies['http'] = os.getenv("proxy", "socks5h://localhost:9150")  # что передается?
    #rs.proxies['https'] = os.getenv("proxy", "socks5h://localhost:9150")
    rs.headers['User-Agent'] = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/53.0.2785.143 Safari/537.36'
    rs.headers['Accept-Language'] = 'Accept-Language: en-US,en;q=0.5'
    result = rs.get(url, timeout=1000)
    rs.close()
    return result
```

```python
def get_rating_yandex(html):
    
    rating = html.find('div', class_ = 'orgpage-header-view__rating-wrapper').find_all('span', class_='business-rating-badge-view__rating-text _size_m')
    if len(rating)!=0:
        rating = rating[0].text
    else:
        rating = "No data"
    return rating
```

```python
def get_rating_google(place_id):
    if place_id is not np.nan:
        try:
            result=gmaps.place(place_id=place_id,fields=['name', 'rating'])
            rating=result['result']['rating']
        except:
            rating = "No data"
    else:
        rating = "No data"
    return rating
```

```python
rating_yandex = []
for url in table['URL_yandex']:
    if url is not np.nan:
        response = get_result(url)
        soup = BeautifulSoup(response.text,'lxml')
        rating_yandex.append(get_rating_yandex(soup))
    else:
        rating_yandex.append("no_data")
```

```python
file = open(r'C:\Users\79192\Desktop\project1\key.txt', 'r')
API_TOKEN= file.read()
file.close()
import googlemaps
gmaps = googlemaps.Client(key=API_TOKEN)
```

```python
rating_google = []
for place_id in table['place_id_google']:
    rating_google.append(get_rating_google(place_id))
```

```python
table['rating_yandex'] = rating_yandex
table['rating_google'] = rating_google
#table['Organization'] =  pd.read_csv(r'C:\\Users\79192\Desktop\project1\data.csv', usecols=['organization'])
```
